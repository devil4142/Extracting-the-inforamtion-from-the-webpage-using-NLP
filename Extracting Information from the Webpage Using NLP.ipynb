```python
import requests 
from bs4 import BeautifulSoup
import spacy 
from spacy.lang.en.stop_words import STOP_WORDS 
from collections import Counter
```

```python
url = 'https://en.wikipedia.org/wiki/CNN'
```

```python
def fetch_webpage_content(url):
    response = requests.get(url)
    return response.text 

content = fetch_webpage_content(url)
print(content)
```

```python
def extract_text(content):
    soup = BeautifulSoup(content, 'html.parser')
    text = ''.join([p.text for p in soup.find_all('p')])
    return text

text = extract_text(content)
print(text)
```

```python
def preprocess_text(text):
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)

    tokens = [token.text.lower() for token in doc if token.is_alpha and token.text.lower() not in STOP_WORDS]
    # Get word frequencies
    word_freq = Counter(tokens)

    # Normalize word frequencies by the most common word
    max_freq = max(word_freq.values()) if word_freq else 1
    word_freq = {word: freq / max_freq for word, freq in word_freq.items()}

    # Create sentences and score them based on word frequencies
    sentences = [sent.text.strip() for sent in doc.sents]
    sentence_scores = {sent: sum(word_freq[word] for word in sent.split() if word in word_freq) for sent in sentences}

    return sentences, sentence_scores

sentences, sentence_scores = preprocess_text(text)
print(sentences)
```

```python
def generate_summary(sentences, sentence_scores, num_sentences=3):
    sorted_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)
    summary = ''.join(sorted_sentences[::num_sentences])
    return summary

summary = generate_summary(sentences, sentence_scores)
print(summary)
```
